```python
# Instalamos la librería de yahoo finance que permite acceder una gran variedad de datos, de entre ellos los precios 
# historicos de diversas acciones
!pip install yfinance
```

```python
# Una vez instalada la importamos al entorno de ejecución
import yfinance as yf
```

```python
# Importamos la librería yfinance y definimos el símbolo de la acción que deseas 
# descargar los datos históricos. En este caso, será "^MXX".
symbol = ["^MXX"]

```

```python
# Importamos datetime que nos ayudara con el formato de fechas
from datetime import datetime, timedelta
# Definimos el rango de fechas de los datos históricos que deseamos descargar. 
# En este caso, se descargarán los datos de los últimos 5 años, 
# por lo que la fecha de inicio será de 5 años antes de la fecha actual
end_date = datetime.today()
start_date = end_date - timedelta(days=5*365)
```

```python
# Creamos un diccionario vacío para almacenar los DataFrames de cada ticker
dataframes = {}

# Iteramos sobre cada ticker en la lista y obtenemos sus datos mensuales
for s in symbol:
    data = yf.Ticker(s).history(start=start_date, end=end_date, interval='1d')
    
# Creamos un DataFrame para el ticker y lo añadimos al diccionario
    dataframes[s] = pd.DataFrame(data)

# Imprimimos los primeros 5 registros de cada DataFrame
for s in symbol:
    print('Ticker:', s)
    print(dataframes[s].head())
```

```python
dataframes["^MXX"].head()
```

```python
mxx = mxx[['Open',	'High',	'Low',	'Close',	'Volume']].to_xml()
mxx = pd.read_xml(mxx)
mxx.head()
```

```python
import datetime
```

```python
mxx['Date']= pd.to_datetime(mxx['Date'])
```

```python
# Importamos más librerías para visualización de datos 
import matplotlib as mpl
import os
import plotly.express as px
```

```python
# Graficamos nuestro precio al cierre
fig = px.line(mxx, x = 'Date',y = 'Close',title = 'Precio al cierre del IPC México')

fig.update_xaxes(
    rangeslider_visible= True,
    rangeselector=dict(
                        buttons = list([
                        dict(count = 1,label = '1m',step='month',stepmode = "backward"),
                        dict(count = 2,label = '6m',step='month',stepmode = "backward"),
                        dict(count = 3,label = '6mYTD',step='month',stepmode = "todate"),
                        dict(count = 4,label = '2y',step='year',stepmode = "backward"),
                        dict(count = 5,label = '3y',step='year',stepmode = "backward"),
                        dict(step= 'all')
                            ])        
                        )
                   )
fig.show()
```

```python
px.area(mxx, x="Date", y= "Close", title = 'Precio al cierre del IPC México')
```
